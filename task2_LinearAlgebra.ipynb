```python
import numpy as np
import matplotlib.pyplot as plt

# Define the coefficient matrix A and the constant matrix B
A = np.array([[1, 3],
              [3, -1],
              [2, 2]])

B = np.array([-2, 4, 1])

# Method 1: Least Squares Method
X_least_squares = np.linalg.lstsq(A, B, rcond=None)[0]

# Method 2: Pseudo-inverse Method
A_pseudo_inv = np.linalg.pinv(A)
X_pseudo_inv = np.dot(A_pseudo_inv, B)

print("Solution using Least Squares Method:", X_least_squares)
print("Solution using Pseudo-inverse Method:", X_pseudo_inv)

# Define the solutions obtained from both methods
x_least_squares = X_least_squares[0]
y_least_squares = X_least_squares[1]

x_pseudo_inv = X_pseudo_inv[0]
y_pseudo_inv = X_pseudo_inv[1]

# Calculate the sum of squared residual errors (SSE) for both solutions
E_least_squares = np.abs(x_least_squares + 3*y_least_squares + 2) + \
                  np.abs(3*x_least_squares - y_least_squares - 4) + \
                  np.abs(2*x_least_squares + 2*y_least_squares - 1)

E_pseudo_inv = np.abs(x_pseudo_inv + 3*y_pseudo_inv + 2) + \
               np.abs(3*x_pseudo_inv - y_pseudo_inv - 4) + \
               np.abs(2*x_pseudo_inv + 2*y_pseudo_inv - 1)

print("SSE using Least Squares Method:", E_least_squares)
print("SSE using Pseudo-inverse Method:", E_pseudo_inv)

# Plotting the solutions
plt.figure(figsize=(10, 8))
plt.scatter(x_least_squares, y_least_squares, color='blue', label='Least Squares Method')
plt.scatter(x_pseudo_inv, y_pseudo_inv, color='red', label='Pseudo-inverse Method')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Solutions Obtained from Least Squares and Pseudo-inverse Methods')
plt.legend()
plt.grid(True)
plt.show()
```
