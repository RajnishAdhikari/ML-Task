```python
import numpy as np
import matplotlib.pyplot as plt

# Function of Euler's method for second-order ODE
def euler_method(omega_n, zeta, y0, v0, t0, tf, dt):
    N = int((tf - t0) / dt)
    t = np.linspace(t0, tf, N+1)
    y = np.zeros(N+1)
    v = np.zeros(N+1)
    y[0] = y0
    v[0] = v0

    for i in range(N):
        y[i+1] = y[i] + dt * v[i]
        v[i+1] = v[i] + dt * (-2 * zeta * omega_n * v[i] - omega_n**2 * y[i])

    return t, y

# Defining constants for the initial conditions
y0 = 1.0  # Initial displacement
v0 = 0.0  # Initial velocity

# Time settings
t0 = 0     # Initial time
tf = 10    # Final time
dt = 0.01  # Time step

# Different scenarios for simulation
params = [
    (2, 0.3),   # Higher damping
    (2, 0.05),  # Lower damping
    (4, 0.1),   # Higher frequency
    (1, 0.1)    # Lower frequency
]

# Dictionary to hold time and displacement data for different scenarios
simulations = {}

# Simulate for each scenario
for omega_n, zeta in params:
    t, y = euler_method(omega_n, zeta, y0, v0, t0, tf, dt)
    simulations[f'n={omega_n}, ={zeta}'] = (t, y)

# Plotting all simulations
plt.figure(figsize=(12, 6))
for label, (t, y) in simulations.items():
    plt.plot(t, y, label=label)

plt.title('Damped Oscillations for Different Parameters')
plt.xlabel('Time')
plt.ylabel('Displacement')
plt.legend()
plt.grid(True)
plt.show()
```

# Parameters of the euler_method function:

* omega_n: The natural frequency of the oscillator. This variable defines the frequency at which the undamped system (no friction or resistance) would oscillate.

* zeta: The damping coefficient. This determines how quickly the oscillations decrease in amplitude. A higher zeta means faster damping.

* y0: The initial displacement of the oscillator from equilibrium. This is where the oscillator starts from at time t0.

* v0: The initial velocity of the oscillator. It represents the speed at which the oscillator is moving at time t0.

* t0: The initial time of the simulation. This is typically set to zero.

* tf: The final time up to which the simulation runs. This defines the length of the time interval over which the system's behavior is simulated.

* dt: The time step used in the simulation. This is the increment between successive time points in the simulation. Smaller values generally yield more accurate results but require more computations.
